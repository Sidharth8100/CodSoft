```python
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
```

```python
from google.colab import files
uploaded = files.upload()  # Manually upload your image file here
```

```python
# Load the image
image_path = list(uploaded.keys())[0]  # Use the uploaded file name
image = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian Blur to reduce noise
gray = cv2.GaussianBlur(gray, (5,5), 0)

# Apply thresholding
gray = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Display the processed image
cv2_imshow(gray)
```

```python
# Run OCR on the processed image
text = pytesseract.image_to_string(gray, config="--psm 6")


# Print the extracted text
print("Extracted Text:\n", text)
```

```python
#Tesseract OCR part trial 2
# Run OCR on the processed image
print("Using psm 6")
text = pytesseract.image_to_string(gray, config="--psm 6")


# Print the extracted text
print("Extracted Text:\n", text)
print("\n \nUsing psm 3")
text = pytesseract.image_to_string(gray, config="--psm 3")
print("Extracted Text:\n", text)

print("\n \nUsing psm 4")
text = pytesseract.image_to_string(gray, config="--psm 4")
print("Extracted Text:\n", text)

#Comment psm 6 -> worst result too much noise found in string
#psm 3 -> many lines left but accuracy was far better than other psms'
```
